```python
import numpy as np
import matplotlib.pyplot as plt
datafile = 'data/iris.txt'
datax = np.loadtxt(datafile,delimiter=',',usecols=(0,1,2,3),unpack=True)
datay = np.loadtxt(datafile,delimiter=',',usecols=(4),unpack=True,dtype='str')
x = np.transpose(datax)
y = np.transpose(datay).reshape(datay.shape[0],1)

#取30条数据作为验证集
testx = np.vstack((x[40:50],x[90:100],x[140:150]))
x = np.vstack((x[0:40],x[50:90],x[100:140]))
print(x.shape[0])
#构建二叉树类
class TreeNode(object):
    def __init__(self,feature,values,nodelist,left=None,right=None,classes = None,array=None,lenth=None):
        self.feature = feature
        self.values = values
        self.nodelist = nodelist
        self.left = left
        self.right = right
        self.classes = classes
        self.array=array
        self.lenth = lenth
    def add(self,lchild):
        self.left = lchild
    def addr(self,rchild):
        self.right = rchild
        
#将字符型类别转换成数字,0,1,2分别代表不同鸢尾花的种类
def transfer_classes(x):
    mask0 = x=='Iris-setosa'
    mask1 = x=='Iris-versicolor'
    mask2 = x=='Iris-virginica'
    x[mask0]='0'
    x[mask1]='1'
    x[mask2]='2'
    x=x.astype(np.float)
    return x
y = transfer_classes(y)
testy = np.vstack((y[40:50],y[90:100],y[140:150]))
y = np.vstack((y[0:40],y[50:90],y[100:140]))
#首先进行CART分类树算法
#对连续值的基尼系数处理，不停的使用二分离散，将值分为两个类别
def continue_compute(x):#这里包含4类特征，鸢尾花数据集
    #分别对四个连续值特征进行排序并且去重
    inside0=[]
    inside1=[]
    inside2=[]
    inside3=[]
    unx0 = np.array(np.unique(x[:,0]))
    unx1 = np.array(np.unique(x[:,1]))
    unx2 = np.array(np.unique(x[:,2]))
    unx3 = np.array(np.unique(x[:,3]))
    #根据以上划分得出二元离散分类点(四个特征分别进行)
    for i in range(unx0.size-1):
        inside0.append(float((unx0[i]+unx0[i+1])/2))
    for i in range(unx1.size-1):
        inside1.append(float((unx1[i]+unx1[i+1])/2))
    for i in range(unx2.size-1):
        inside2.append(float((unx2[i]+unx2[i+1])/2))
    for i in range(unx3.size-1):
        inside3.append(float((unx3[i]+unx3[i+1])/2))
    return inside0,inside1,inside2,inside3
#x.sort(axis=0)
i0,i1,i2,i3=continue_compute(x)#得到了二元离散划分点的值

#假设选定二元离散划分点为p0，p1，p2，p3，计算它们的基尼系数
def compute_gini(x,p0,n,y):#p0为划分点，n为第n个特征
    #首先针对第一个特征,计算他的基尼指数，作为二元离散划分点，就只有一个划分点
    x=np.hstack((x,y))
    mask0_less = x[:,n-1]<p0
    mask0_big = x[:,n-1]>p0
    less0 =x[mask0_less].shape[0]
    big0 = x[mask0_big].shape[0]
    num0_1=0
    num0_2=0
    num0_3=0
    for i in range(x[mask0_less].shape[0]):
        if(x[mask0_less][i,4]==0):
            num0_1 = num0_1+1
        elif(x[mask0_less][i,4]==1):
            num0_2 = num0_2+1
        elif(x[mask0_less][i,4]==2):
            num0_3 = num0_3+1
    bnum0_1=0
    bnum0_2=0
    bnum0_3=0
    for i in range(x[mask0_big].shape[0]):
        if(x[mask0_big][i,4]==0):
            bnum0_1 = bnum0_1+1
        elif(x[mask0_big][i,4]==1):
            bnum0_2 = bnum0_2+1
        elif(x[mask0_big][i,4]==2):
            bnum0_3 = bnum0_3+1
    left = float((less0/x.shape[0])*(1-((num0_1/less0)**2+(num0_2/less0)**2+(num0_3/less0)**2)))
    right = float((big0/x.shape[0])*(1-((bnum0_1/big0)**2+(bnum0_2/big0)**2+(bnum0_3/big0)**2)))
    gini = left+right
    return gini#得到第n个特征值，使用p0为划分点的基尼指数

#选取基尼系数最小的划分点进行每一个特征划分
def Gini_select(x,l0,l1,l2,l3,y):#总共四个特征
    minGini0= 100
    point = [0,0,0,0]#记录四个特征的划分点
    for i in range(len(l0)):#计算每一个划分点的基尼系数针对第一个特征
        if(minGini0>=compute_gini(x,l0[i],1,y)):
            minGini0 = compute_gini(x,l0[i],1,y)
            point[0]=l0[i]
    minGini1= 100
    for i in range(len(l1)):#计算每一个划分点的基尼系数针对第二个特征
        if(minGini1>=compute_gini(x,l1[i],2,y)):
            minGini1 = compute_gini(x,l1[i],2,y)
            point[1]=l1[i]
    minGini2= 100
    for i in range(len(l2)):#计算每一个划分点的基尼系数针对第三个特征
        if(minGini2>=compute_gini(x,l2[i],3,y)):
            minGini2 = compute_gini(x,l2[i],3,y)
            point[2]=l2[i]
    minGini3= 100
    for i in range(len(l3)):#计算每一个划分点的基尼系数针对第四个特征
        if(minGini3>=compute_gini(x,l3[i],4,y)):
            minGini3 = compute_gini(x,l3[i],4,y)
            point[3]=l3[i]
    #print(minGini0)
    #print(minGini1)
   # print(minGini2)
    #print(minGini3)
   #得到最佳划分点进行四个特征点的划分
    #选择返回哪一个基尼系数最小的特征用于划分
    if((minGini0<=minGini1)&(minGini0<=minGini2)&(minGini0<=minGini3)):
        return point,minGini0,1
    elif((minGini1<=minGini2)&(minGini1<=minGini3)):
        return point,minGini1,2
    elif((minGini2<=minGini3)):
        return point,minGini2,3
    else:
        return point,minGini3,4
    return 0
   
p,minG,f= Gini_select(x,i0,i1,i2,i3,y)#得到四个最佳划分点
#print(p)
#print(f)

#根据划分点，以及第几个特征进行拆分集合x,返回拆分的x1，x2
def make_up_point(x,p0,n,y):
    x=np.hstack((x,y))
    mask_less = x[:,n-1]<p0
    mask_big = x[:,n-1]>p0
    return x[mask_less],x[mask_big]
tree = []
feature = []
Gini = []
node = []
treenode = []

#经过前期的准备工作，准备构建Cart分类树
def judge_equal(x):
    for i in range(x.shape[0]-1):
        if(x[i]!=x[i+1]):
            return 0
    
root = TreeNode(f,p[f-1],x,lenth=x.shape[0])

def Cart(x,y,p,f,minG,root):
    #根据前面的划分，选取第三个特征做划分
    tree.append(p[f-1])#构建根节点,记录根据第几个特征进行划分，同时划分点的大小是多少
    feature.append(f)
    Gini.append(minG)
    #进行划分，得到新的集合
    x1,x2=make_up_point(x,p[f-1],f,y)
    node.append(x1)
    node.append(x2)
    #创建根节点
    #print(x1.shape)
    #print(x2.shape)
    if(judge_equal(x1[:,-1::]))==0:
        i0_1,i1_1,i2_1,i3_1 = continue_compute(x1[:,0:4])
        p_1,minG_1,f_1 =Gini_select(x1[:,0:4],i0_1,i1_1,i2_1,i3_1,x1[:,-1::]) 
        lchild = TreeNode(f_1,p_1[f_1-1],x1,lenth=x1.shape[0])
        treenode.append(lchild)
        root.left=lchild
        Cart(x1[:,0:4],x1[:,-1::],p_1,f_1,minG_1,lchild)
    else:
        i0_1,i1_1,i2_1,i3_1 = continue_compute(x1[:,0:4])
        p_1,minG_1,f_1 =Gini_select(x1[:,0:4],i0_1,i1_1,i2_1,i3_1,x1[:,-1::]) 
        lchild = TreeNode(0,0,x1,classes=np.mean(x1[:,-1::]),lenth=x1.shape[0])
        treenode.append(lchild)
        root.left=lchild
    if(judge_equal(x2[:,-1::]))==0:
        i0_2,i1_2,i2_2,i3_2 = continue_compute(x2[:,0:4])
        p_2,minG_2,f_2 =Gini_select(x2[:,0:4],i0_2,i1_2,i2_2,i3_2,x2[:,-1::])
        rchild = TreeNode(f_2,p_2[f_2-1],x2,lenth=x2.shape[0])
        treenode.append(rchild)
        root.right=rchild
        Cart(x2[:,0:4],x2[:,-1::],p_2,f_2,minG_2,rchild)
    else:
        i0_2,i1_2,i2_2,i3_2 = continue_compute(x2[:,0:4])
        p_2,minG_2,f_2 =Gini_select(x2[:,0:4],i0_2,i1_2,i2_2,i3_2,x2[:,-1::])
        rchild = TreeNode(0,0,x2,classes=np.mean(x2[:,-1::]),lenth=x2.shape[0])
        treenode.append(rchild)
        root.right=rchild
treenode.append(root)

Cart(x,y,p,f,minG,root)
#print(tree)
#print(feature)

#for i in range(len(node)):
    #print(len(node[i]))
list_node = []
list_node_without_array = []
#为实现决策树可视化进行转换字典
def dict_tree0(nodelist,dict2,num):
    if(nodelist!=None):
        dict1 = dict(number = num,feature=nodelist.feature,values = nodelist.values,lenth = nodelist.nodelist.shape[0],classes=nodelist.classes,node=nodelist.nodelist)
        dict3 = dict(number = num,feature=nodelist.feature,values = nodelist.values,lenth = nodelist.nodelist.shape[0],classes=nodelist.classes)
        lenth =nodelist.nodelist.shape[0]
        dict2=dict(Node=dict1)
        dict4=dict(Node=dict3)
        if(nodelist.left!=None):
            dict_tree0(nodelist.left,dict2,2*num)
        if(nodelist.right!=None):
            dict_tree0(nodelist.right,dict2,2*num+1)
        #print('节点序号：'+str(num))
        list_node.append(dict2)
        list_node_without_array.append(dict4)
        #print(dict2)
    else:
        return 0
    
dict3 = {}
final = dict_tree0(root,dict3,1)
#将获得的二叉决策树根据序号大小进行排序
def sort_DT(list_node):
    for j in range(len(list_node)-1,-1,-1):
        for i in range(0,j):
            if(list_node[i]['Node']['number']>list_node[i+1]['Node']['number']):
                temp = list_node[i+1]
                list_node[i+1] = list_node[i]
                list_node[i]=temp
    #for k in range(len(list_node)):
        #print(list_node[k])
sort_DT(list_node_without_array)
sort_DT(list_node)
#print(len(list_node))
#print(list_node)
#验证没有

#没有经过剪枝的Cart决策树，进行分类，得到的分类结果

#进行决策树画图\
#定义文本框和箭头格式
decisionNode = dict(boxstyle='sawtooth',fc='0.8')
leafNode = dict(boxstyle = 'round4',fc='0.8')
arrow_args = dict(arrowstyle ='<-')

#绘制带箭头的注解
def plotNode(nodeTxt,centerPt,parentPt,nodeType):
    createPlot.ax1.annotate(nodeTxt,xy=parentPt,xycoords = 'axes fraction',xytext=centerPt,textcoords='axes fraction',va='center',ha='center',bbox=nodeType,arrowprops = arrow_args)
def createPlot():
    fig = plt.figure(1,facecolor='white')
    fig.clf()
    createPlot.ax1  = plt.subplot(111,frameon=False)
   # plotNode(list_node[0]['Node']['feature'],(6,8),(5,9),decisionNode)
    plotNode('a decision node',(0.5,0.1),(0.1,0.5),decisionNode)
    plotNode('a leaf node',(0.8,0.1),(0.3,0.8),leafNode)
    plt.show()
#得到决策分类树
#def get_cart_tree(x):
    
#createPlot()
   # mask0 = x=='Iris-setosa'
    #mask1 = x=='Iris-versicolor'
    #mask2 = x=='Iris-virginica' 
#针对一个输入向量x，对其经过决策树，得到类别
def predict_0(x,node,res):#针对这个未经过剪枝的决策树所编写的判定函数,这个判定函数需要递归
    #首先判断该节点是否为空
    if(node==None):
        return 0
    else:
    #首先根据结点进行判断，看该结点是否是叶节点，若为叶节点则直接输出类别
        if(node.classes==None):#如果不是叶节点
            #首先判断该节点的决策内容
            if(float(node.values)<=float(x[node.feature-1])):
                #print('big')
                #则继续判断其右孩子
                predict_0(x,node.right,res)
            else:
                #判断其左孩子
                #print('small')
                predict_0(x,node.left,res)
        else:
            #print('122')
            res.append(int(node.classes))

#编写测试用例
#xtest1 = [6,3,4.8,2]
#predict_0(xtest1,root)
#print(res)

def comput_currency(realy,realx,node):
    res = [] 
    for i in range(realx.shape[0]):
        predict_0(realx[i],node,res)
    rate = 0
    print(len(res))
    for j in range(realy.shape[0]):
        if(res[j]==realy[j]):
            rate = rate+1
    print("验证集在CART决策树上的分类正确率为:{:.2f}%".format((rate/realy.shape[0])*100))
    return 0
#通过测试样例，发现预测函数与决策树分类符合，
#继续进行剪枝工作

#将treenode与list_node进行匹配
#treenode相当于一个层次遍历，使用队列
def fit(root):
    outList=[]
    queue = [root]
    while queue!=[] and root:
        outList.append(queue[0])
        if(queue[0].left!=None):
            queue.append(queue[0].left)
        if(queue[0].right!=None):
            queue.append(queue[0].right)
        queue.pop(0)
    return outList
treenode = fit(root)

#首先进行后剪枝(post-purning),对付过拟合手段
def post_purning(list_node,treenode):
    #考虑序号最后一个结点,肯定为叶结点
    #得到其父结点的分类信息
    father_number = int(list_node[-1]['Node']['number']/2)
    print('父亲结点序号',father_number)
    #然后找出其父结点的所有孩子节点，找出其中包含集合大小以及类别信息
    for i in range(len(list_node)):
        if(int(list_node[i]['Node']['number']))==father_number:
            father_node_num = i
        if(int(list_node[i]['Node']['number']))==int(father_number*2):
            lchild_node_num = i
        if(int(list_node[i]['Node']['number']))==int(father_number*2)+1:
            rchild_node_num = i
    print('父亲结点索引',father_node_num)
    print('左孩子索引',lchild_node_num)
    print('右孩子索引',rchild_node_num)
    #找出其孩子结点,比较孩子结点中包含的训练集大小
    print('左孩子大小',treenode[father_node_num].left.lenth)
    print('右孩子大小',treenode[father_node_num].right.lenth)
    if(treenode[father_node_num].left.lenth<treenode[father_node_num].right.lenth):
        #如果右孩子中类别训练集数目比左孩子多，则让父节点中的类别沿用右孩子的类别
        print('right')
        treenode[father_node_num].classes = treenode[father_node_num].right.classes
        list_node[father_node_num]['Node']['classes']=list_node[rchild_node_num]['Node']['classes']
    else:
        print('left')
        treenode[father_node_num].classes = treenode[father_node_num].left.classes
        list_node[father_node_num]['Node']['classes']=list_node[lchild_node_num]['Node']['classes']
    
    treenode[father_node_num].feature = 0#将父结点转换为叶结点
    treenode[father_node_num].values = 0
    treenode[father_node_num].lenth = int(treenode[lchild_node_num].lenth)+int(treenode[rchild_node_num].lenth)
    treenode[father_node_num].left = None
    treenode[father_node_num].right = None
        #删除结点孩子信息
    #right = list_node[rchild_node_num]['Node']['node']
    #left = list_node[lchild_node_num]['Node']['node']
    #list_node[father_node_num]['Node']['node'] = np.vstack((left,right))
    list_node[father_node_num]['Node']['feature']=0
    list_node[father_node_num]['Node']['values']=0
    list_node[father_node_num]['Node']['lenth']=int(list_node[lchild_node_num]['Node']['lenth'])+int(list_node[rchild_node_num]['Node']['lenth'])
    del list_node[rchild_node_num]
    del list_node[lchild_node_num]
    del treenode[rchild_node_num]
    del treenode[lchild_node_num]


def print_list(listnode):
    for i in range(len(listnode)):
        print('序号：',listnode[i]['Node']['number'])
        print('feature:',listnode[i]['Node']['feature'])
        print('values:',listnode[i]['Node']['values'])
        print('classes:',listnode[i]['Node']['classes'])
        print('lenth:',listnode[i]['Node']['lenth'])
def print_tree(tree):
    for i in range(len(tree)):
        print('序号',i)
        print('feature',tree[i].feature)
        print('value',tree[i].values)
        print('lenth',tree[i].lenth)
        print('lchild',tree[i].left)
        print('rchild',tree[i].right)
        print('类别',tree[i].classes)

#进行预剪枝操作，对付过拟合的方法
def pre_purning():
    return 0

#print_list(list_node)
#未经过剪枝的决策树分类结果
comput_currency(testy,testx,root)
print('第一次后剪枝')
post_purning(list_node,treenode)

#计算后剪枝后新的准确度
comput_currency(testy,testx,root)
print(len(list_node))
print(len(treenode))


print('第二次后剪枝')
#继续进行后剪枝
#print(list_node)
post_purning(list_node,treenode)
comput_currency(testy,testx,root)
print(len(list_node))
print(len(treenode))

print('第三次后剪枝')
#继续进行后剪枝
#print(treenode[6].right)
#print(list_node)
#print_tree(treenode)
post_purning(list_node,treenode)
comput_currency(testy,testx,root)
print(len(list_node))
print(len(treenode))



print('第四次后剪枝')
#继续进行后剪枝
#print(treenode[6].right)
#print(list_node)
#print_tree(treenode)
post_purning(list_node,treenode)
comput_currency(testy,testx,root)
print(len(list_node))
print(len(treenode))
print_list(list_node)

print('第五次后剪枝')
#继续进行后剪枝
#print(treenode[6].right)
#print(list_node)
#print_tree(treenode)
post_purning(list_node,treenode)
comput_currency(testy,testx,root)
print(len(list_node))
print(len(treenode))
print('第六次后剪枝')
#继续进行后剪枝
#print(treenode[6].right)
#print(list_node)
#print_tree(treenode)
post_purning(list_node,treenode)
comput_currency(testy,testx,root)
print(len(list_node))
print(len(treenode))
#print_tree(treenode)
#print(np.array(list_node[3]['Node']['node']).shape)
#print('第七次后剪枝')
#继续进行后剪枝
#print(treenode[6].right)
#print(list_node)
#print_tree(treenode)
post_purning(list_node,treenode)
comput_currency(testy,testx,root)

import os
cmd = "dot  5.gv -Tpng -o 5.png"
os.system(cmd)
```
