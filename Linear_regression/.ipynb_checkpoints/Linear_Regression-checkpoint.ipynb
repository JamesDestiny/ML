```python
import numpy as np
import matplotlib.pyplot as plt
#
datafile = 'data/ex1data1.txt'
cols = np.loadtxt(datafile,delimiter=',',usecols=(0,1),unpack=True)#delimters
#unpack,1xM
x=np.transpose(np.array(cols[0:1]))#1XMMx1
y=np.transpose(np.array(cols[1:2]))#1XMMx1
x=np.insert(x,0,1,axis=1)#xaxis=11xMx2
m=x.shape[0]#M

#
plt.figure(figsize=(9,6),facecolor='w')
plt.xlabel('x')
plt.ylabel('y')
plt.title('dataset plot')
plt.plot(x[:,1],y[:,0],'rx')

#
def hypothesis(theta,X):
    #
    #h(x)=X*theta
    return np.dot(X,theta)#

def CostFuction(theta,X,Y):
    #J(theta)
    #thetan
    J_theta= float((1/(2*m))*(np.dot(((hypothesis(theta,X)-Y).T),(hypothesis(theta,X)-Y))))
    e_theta = float((1./(2*m)) * np.dot((hypothesis(theta,X)-Y).T,(hypothesis(theta,X)-Y)))
    return J_theta

#gradient descentalpha iterations
alpha = 0.01
iterations = 3000

#theta,thetan
init_theta = np.zeros((x.shape[1],1))#n

#
def Gradient_descent(theta,X,Y):
    CostList_with_different_theta=[]#theta
    theta_List=[]#theta
    #repeat,
    for nums in range(iterations):
        temp_theta = theta
        #thetaLIST
        CostList_with_different_theta.append(CostFuction(temp_theta,X,Y))
        #thetaList
        theta_List.append(temp_theta)
        #theta
        for j in range(len(temp_theta)):#
            temp_theta[j]=theta[j]-alpha*(1/m)*np.sum((hypothesis(theta,X)-Y)*(np.array(X[:,j]).reshape(m,1)))
        theta = temp_theta#theta
    return theta,theta_List,CostList_with_different_theta
theta,thetaList,CostList= Gradient_descent(init_theta,x,y)                                                    

#thetatheta
#
def plotCostFuction(CostList):
    plt.figure(figsize=(10,6))
    plt.plot(range(len(CostList)),CostList,'ro')
    plt.grid(True)
    plt.title('costfuction in different iterations')
    plt.xlabel("iterations")
    plt.ylabel("cost fuction values")
    plt.xlim([-0.05*iterations,1.05*iterations])#x
    plt.ylim([3,7])
    plt.plot
plotCostFuction(CostList)

#thetatheta,
def Linear_model_with_theta(theta,xi):
    model_predict = theta[0]+theta[1]*xi#theta
    return model_predict

#
def plotPredict():
    plt.figure(figsize=(10,6))
    plt.grid(False)
    plt.plot(x[:,1],y[:,0],'r*',label="training data")
    plt.plot(x[:,1],Linear_model_with_theta(theta,x[:,1]),'b-',label='Predict result')
    plt.title("hypothesis h(x)=%0.2f+%0.2fx"%(theta[0],theta[1]))
    plt.legend()
plotPredict()
CostList[-1]
```

```python
#J(theta)
#3d,theta
from mpl_toolkits.mplot3d import axes3d,Axes3D
from matplotlib import cm
import itertools

def plot_3D_J_Fuction():
    figure = plt.figure(figsize=(12,12))
    ax=figure.gca(projection='3d')
    xvals = np.arange(-10,10,.5)
    yvals = np.arange(-1,4,.1)
    xs, ys, zs = [],[],[]
    for rx in xvals:
        for ry in yvals:
            xs.append(rx)
            ys.append(ry)
            zs.append(CostFuction(np.array([[rx],[ry]]),x,y))
    scat = ax.scatter(xs,ys,zs,c=np.abs(zs),cmap=plt.get_cmap('YlOrRd'))
    plt.xlabel(r'$\theta_0$',fontsize=30)                         
    plt.ylabel(r'$\theta_1$',fontsize=30)
    plt.title('Cost (Minimization Path Shown in Blue)',fontsize=30)
    plt.plot([x[0] for x in thetaList],[x[1] for x in thetaList],CostList,'bo-')
    plt.grid(False)
    plt.show()
plot_3D_J_Fuction()
```
