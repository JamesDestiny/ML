```python
import numpy as np
import matplotlib.pyplot as plt
#读取数据
datafile = 'data/ex1data1.txt'
cols = np.loadtxt(datafile,delimiter=',',usecols=(0,1),unpack=True)#delimters使用间隔符
#unpack参数，把每一列当作向量输出,而且输出的每一列数据都变为行向量，即1xM
x=np.transpose(np.array(cols[0:1]))#找出第一行1XM，转置变为Mx1
y=np.transpose(np.array(cols[1:2]))#找出第二行1XM，转置变为Mx1
x=np.insert(x,0,1,axis=1)#在x中的第一列（参数axis=1）中插入常数列全为1，此时x为Mx2矩阵
m=x.shape[0]#数据项的个数，为M

#初步查看数据，观察其规律
plt.figure(figsize=(9,6),facecolor='w')
plt.xlabel('x')
plt.ylabel('y')
plt.title('dataset plot')
plt.plot(x[:,1],y[:,0],'rx')

#定义假设函数与损失函数
def hypothesis(theta,X):
    #假设函数
    #h(x)=X*theta
    return np.dot(X,theta)#

def CostFuction(theta,X,Y):
    #损失函数J(theta)
    #theta是一个n维列向量，
    J_theta= float((1/(2*m))*(np.dot(((hypothesis(theta,X)-Y).T),(hypothesis(theta,X)-Y))))
    e_theta = float((1./(2*m)) * np.dot((hypothesis(theta,X)-Y).T,(hypothesis(theta,X)-Y)))
    return J_theta

#定义梯度下降gradient descent的学习率alpha ，和迭代次数iterations
alpha = 0.01
iterations = 3000

#初始化theta,注意theta是一个n维列向量
init_theta = np.zeros((x.shape[1],1))#设置为n行一列

#定义梯度下降函数
def Gradient_descent(theta,X,Y):
    CostList_with_different_theta=[]#记录沿梯度下降过程中不同theta下的不同损失函数值
    theta_List=[]#记录梯度下降过程中不同的theta值
    #进行迭代repeat,初始化迭代次数前面已经定义
    for nums in range(iterations):
        temp_theta = theta
        #将新的theta的损失函数值LIST中
        CostList_with_different_theta.append(CostFuction(temp_theta,X,Y))
        #将新的theta值加入List中
        theta_List.append(temp_theta)
        #每一次更新theta的值
        for j in range(len(temp_theta)):#矩阵点乘，要求两个矩阵必须同样大小
            temp_theta[j]=theta[j]-alpha*(1/m)*np.sum((hypothesis(theta,X)-Y)*(np.array(X[:,j]).reshape(m,1)))
        theta = temp_theta#更新theta的值
    return theta,theta_List,CostList_with_different_theta
theta,thetaList,CostList= Gradient_descent(init_theta,x,y)                                                    

#根据得到的theta，以及梯度下降过程中theta历史值与相对应的损失函数值，来进行数据分析
#显示损失函数
def plotCostFuction(CostList):
    plt.figure(figsize=(10,6))
    plt.plot(range(len(CostList)),CostList,'ro')
    plt.grid(True)
    plt.title('costfuction in different iterations')
    plt.xlabel("iterations")
    plt.ylabel("cost fuction values")
    plt.xlim([-0.05*iterations,1.05*iterations])#设置x轴的取值范围
    plt.ylim([3,7])
    plt.plot
plotCostFuction(CostList)

#根据上文中的梯度下降得到的最终的theta值，根据该theta值拟合出自己的模型,
def Linear_model_with_theta(theta,xi):
    model_predict = theta[0]+theta[1]*xi#包含两个参数theta的单变量线性模型
    return model_predict

#对比预测值与原数据值
def plotPredict():
    plt.figure(figsize=(10,6))
    plt.grid(False)
    plt.plot(x[:,1],y[:,0],'r*',label="training data")
    plt.plot(x[:,1],Linear_model_with_theta(theta,x[:,1]),'b-',label='Predict result')
    plt.title("hypothesis h(x)=%0.2f+%0.2fx"%(theta[0],theta[1]))
    plt.legend()
plotPredict()
CostList[-1]
```

```python
#分析代价函数J(theta)
#导入3d图像库,观察代价函数在不同theta值下进行梯度下降的效果图
from mpl_toolkits.mplot3d import axes3d,Axes3D
from matplotlib import cm
import itertools

def plot_3D_J_Fuction():
    figure = plt.figure(figsize=(12,12))
    ax=figure.gca(projection='3d')
    xvals = np.arange(-10,10,.5)
    yvals = np.arange(-1,4,.1)
    xs, ys, zs = [],[],[]
    for rx in xvals:
        for ry in yvals:
            xs.append(rx)
            ys.append(ry)
            zs.append(CostFuction(np.array([[rx],[ry]]),x,y))
    scat = ax.scatter(xs,ys,zs,c=np.abs(zs),cmap=plt.get_cmap('YlOrRd'))
    plt.xlabel(r'$\theta_0$',fontsize=30)                         
    plt.ylabel(r'$\theta_1$',fontsize=30)
    plt.title('Cost (Minimization Path Shown in Blue)',fontsize=30)
    plt.plot([x[0] for x in thetaList],[x[1] for x in thetaList],CostList,'bo-')
    plt.grid(False)
    plt.show()
plot_3D_J_Fuction()
```

```python
#第二种方法，使用正规方程，进行求解theta
def Nomar_equation(x,y,theta):
    return np.linalg.solve(np.dot(x.T,x),np.dot(x.T,y))
#先进行判断X乘X的转置是否可逆
if(np.linalg.det(np.dot(x.T,x))==0):
    print("矩阵不可逆")
else:
    final_theta=Nomar_equation(x,y,init_theta)
#得到通过Nomal_equation得到的theta值，与训练数据进行对比
def Linear_Model_With_Nomal_equation(final_theta,xi):
    return final_theta[0]+final_theta[1]*xi
#计算正规方程下得出的theta值对应的损失函数值
def compute_Ne_theta_costValues(final_theta):
    return CostFuction(final_theta,x,y)
#通过绘图展示出来
def plot_Nomal_equation_compare_with_data():  
    plt.figure(figsize=(10,6),num=3)
    plt.grid(True)
    plt.plot(x[:,1],y[:,0],'r*',label="training data")
    plt.plot(x[:,1],Linear_model_with_theta(theta,x[:,1]),'b-',label='Predict result by gradient descent')
    plt.plot(x[:,1],Linear_Model_With_Nomal_equation(final_theta,x[:,1]),'g-',label='Predict result by nomal equation')
    plt.title("hypothesis h(x)=%0.2f+%0.2fx VS h(x)=%0.2f+%0.2fx"%(theta[0],theta[1],final_theta[0],final_theta[1]))
    plt.legend()
plot_Nomal_equation_compare_with_data()
print("两种minimize损失函数的方法，梯度下降VS正规方程，损失大小结果分别为为%.10f VS %.10f"%(CostList[-1],CostFuction(final_theta,x,y)))
```
